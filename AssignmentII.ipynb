# Q3 Distribution üìä

Q.IV
1. Generate a random sample of size 100 from poisson distribution and check its normality
2. Generate a random sample of size 100 from normal distribution and check its normality

```python
import numpy as np
import random
import scipy as sp
from scipy.stats import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#Generate a random sample of size 100 from poisson distribution and check its normality
sample = np.random.poisson(lam=10, size=100)
plt.hist(sample,bins=20,edgecolor='black')
plt.title('Histogram of Random Sample from Poisson Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
## making qq plot to check if it lines with normal
## using sp.stats.probplot to check normality remember to use scipy.probplot
sp.stats.probplot(sample,dist="norm",plot=plt)
plt.title("QQ Plot")
plt.show()
```

```python
# Generate a random sample of size 100 from normal distribution and check its normality
#sample=np.random.normal(loc=0,scale=1,size=100) for standard normal distribution
sample=np.random.normal(loc=4,scale=3,size=100)
plt.hist(sample,bins=20,edgecolor='red')
plt.title('Histogram of Random Sample from Normal Distribution')
plt.show()
sp.stats.probplot(sample,dist="norm",plot=plt)
plt.title("QQ Plot")
plt.show()
```
